```python
# 토큰화
tokenizer = Okt()
tokenized = []
for i in range(0, len(newsIT['content'])):
  sentence = newsIT['content'][i]
  tk = tokenizer.pos(sentence)
  tokenized.append(tk)

newsIT['token_okt'] = tokenized

newsIT
```

```python
# stopwords 중복 제거
stopwords=['대한', '통해', '이번', '위해', '지난', '사용', '올해', '기자', '서비스', '기술', '사업', '기업', '연합뉴스', '승진', '위원장', '회장']
stopwords=list(set(stopwords))
```

```python
# 토큰화 된 본문이 (word, tag)의 꼴을 취한다.
# 필요한 단어만 남기려면 본문의 word는 불용어 사전에 있으면 안되고, tag가 우리가 원하는 tag인 단어를 찾아야한다.

cleans = []
for i in range(len(newsIT)):
    clean_words = [] 
    sentence = newsIT['token_okt'][i]
    for word, pos in sentence:
      if word not in stopwords and pos in ('Noun', 'Adjective', 'Adverb'):# kr_wordrank에서는 'Punctuation' 구두점 추가
        clean_words.append(word)
    cleans.append(clean_words)

# 위 작업을 거친 필요한 단어들을 새로운 칼럼에 'word'만 저장한다 ('word'만 저장하는 이유는 토픽모델링에서 그렇게 써야하기 때문이다)
newsIT['clean_words'] = cleans

newsIT
```

```python
doclist = []
for i in range(len(newsIT)):
  nouns = ' '.join(newsIT['clean_words'][i])
  doclist.append(nouns)

# 역시 수행 결과를 새로운 칼럼에 저장
newsIT['clean_as_sentence'] = doclist
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features = 1000)
x = tfidf_vectorizer.fit(newsIT.clean_as_sentence)
tfidf_df = pd.DataFrame(x.transform(newsIT.clean_as_sentence).toarray(), columns = sorted(tfidf_vectorizer.vocabulary_))

important_words = []
for i in range(0, len(tfidf_df)):
  a = []
  imp_word = tfidf_df.iloc[i, :].index[tfidf_df.iloc[i, :].values > 0.1].tolist()
  a.append(imp_word)
  important_words.append(a)

newsIT['important_words'] = important_words
# 키워드 추출 끝
```

```python
text1 = newsIT['important_words'].to_list()
tw = Twitter()

text = ''.join(str(v) for v in text1)
text_nouns = FreqDist(tw.nouns(text))

wordcloud = WordCloud(
    font_path ='NanumBarunGothic',
    width = 800,
    height = 800,
    background_color="white"
)

wordcloud = wordcloud.generate_from_frequencies(text_nouns)

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
